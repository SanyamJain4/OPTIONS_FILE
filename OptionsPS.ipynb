Handling missing values using interpolation

```python
import pandas as pd
import numpy as np

# Define the function to handle NaN values in numeric columns
def fill_nan_with_iqr(df, columns):
    for col in columns:
        if pd.api.types.is_numeric_dtype(df[col]):  # Only process numeric columns
            if df[col].isnull().sum() > 0:  # Check if there are missing values
                # Calculate Q1, Q3, and IQR
                Q1 = df[col].quantile(0.25)
                Q3 = df[col].quantile(0.75)
                IQR = Q3 - Q1
                lower_bound = Q1 - 1.5 * IQR
                upper_bound = Q3 + 1.5 * IQR

                # Interpolate missing values
                df[col] = df[col].interpolate(method='linear', limit_direction='both')

                # Clip values to the IQR range
                df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
    return df

# Identify numeric columns (excluding 'Date  ' and 'Expiry  ')
numeric_columns_put = put_option_data.select_dtypes(include=[np.number]).columns
numeric_columns_call = call_option_data.select_dtypes(include=[np.number]).columns

# Apply the function to the DataFrames
put_option_data = fill_nan_with_iqr(put_option_data, numeric_columns_put)
call_option_data = fill_nan_with_iqr(call_option_data, numeric_columns_call)
```

Using technical Indicators For Price Movements

```python

# Calculate SMA and EMA
put_option_data['SMA_10'] = put_option_data['Close'].rolling(window=10).mean()
put_option_data['SMA_50'] = put_option_data['Close'].rolling(window=50).mean()
put_option_data['EMA_10'] = put_option_data['Close'].ewm(span=10, adjust=False).mean()
put_option_data['EMA_50'] = put_option_data['Close'].ewm(span=50, adjust=False).mean()
```

```python
call_option_data['SMA_10']=call_option_data['Close'].rolling(window=10).mean()
call_option_data["SMA_50"]=call_option_data["Close"].rolling(window=50).mean()
call_option_data["EMA_10"]=call_option_data["Close"].ewm(span=10,adjust=False).mean()
call_option_data["EMA_50"]=call_option_data["Close"].ewm(span=50,adjust=False).mean()
```

```python
# Calculate RSI
delta = put_option_data['Close'].diff()
gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
rs = gain / loss
put_option_data['RSI'] = 100 - (100 / (1 + rs))
```

```python
# Calculate RSI
delta = call_option_data['Close'].diff()
gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
rs = gain / loss
call_option_data['RSI'] = 100 - (100 / (1 + rs))
```

```python
put_option_data['Signal'] = 'Hold'
put_option_data.loc[put_option_data['SMA_10'] > put_option_data['SMA_50'], 'Signal'] = 'Buy'
put_option_data.loc[put_option_data['RSI'] > 70, 'Signal'] = 'Sell'
put_option_data.loc[put_option_data['RSI'] < 30, 'Signal'] = 'Buy'
```

```python
call_option_data['Signal'] = 'Hold'
call_option_data.loc[call_option_data['SMA_10'] > call_option_data['SMA_50'], 'Signal'] = 'Buy'
call_option_data.loc[call_option_data['RSI'] > 70, 'Signal'] = 'Sell'
call_option_data.loc[call_option_data['RSI'] < 30, 'Signal'] = 'Buy'
```

Implementing Option Strategy

```python
bull_put_spread = (put_option_data['SMA_10'] < put_option_data['SMA_50']) & (put_option_data['RSI'] > 70)
iron_butterfly = (abs(put_option_data['SMA_10'] - put_option_data['SMA_50']) < 0.5) & (put_option_data['RSI'] == 50)
long_put = (put_option_data['SMA_10'] < put_option_data['SMA_50']) & (put_option_data['RSI'].between(30, 70))
straddle = (abs(put_option_data['SMA_10'] - put_option_data['SMA_50']) < 0.2) & (put_option_data['RSI'].between(45, 55))
strangle = (abs(put_option_data['SMA_10'] - put_option_data['SMA_50']) < 1.0) & (put_option_data['RSI'].between(45, 55))

# Assign strategies for Put Option
put_option_data.loc[bull_put_spread, 'Option_Strategy'] = 'Bear Put Spread'
put_option_data.loc[iron_butterfly, 'Option_Strategy'] = 'Iron Butterfly'
put_option_data.loc[long_put, 'Option_Strategy'] = 'Long Put'
put_option_data.loc[straddle, 'Option_Strategy'] = 'Straddle'
put_option_data.loc[strangle, 'Option_Strategy'] = 'Strangle'

# Implementing Option Strategy for Call Option
call_option_data['Option_Strategy'] = None

# Define refined strategy conditions for Call Option
bull_call_spread = (call_option_data['SMA_10'] > call_option_data['SMA_50']) & (call_option_data['RSI'] < 40)
iron_condor = (abs(call_option_data['SMA_10'] - call_option_data['SMA_50']) < 0.5) & (call_option_data['RSI'].between(40, 60))
long_call = (call_option_data['SMA_10'] > call_option_data['SMA_50']) & (call_option_data['RSI'].between(30, 70))
straddle = (abs(call_option_data['SMA_10'] - call_option_data['SMA_50']) < 0.2) & (call_option_data['RSI'].between(45, 55))
strangle = (abs(call_option_data['SMA_10'] - call_option_data['SMA_50']) < 1.0) & (call_option_data['RSI'].between(45, 55))

# Assign strategies for Call Option
call_option_data.loc[bull_call_spread, 'Option_Strategy'] = 'Bull Call Spread'
call_option_data.loc[iron_condor, 'Option_Strategy'] = 'Iron Condor'
call_option_data.loc[long_call, 'Option_Strategy'] = 'Long Call'
call_option_data.loc[straddle, 'Option_Strategy'] = 'Straddle'
call_option_data.loc[strangle, 'Option_Strategy'] = 'Strangle'

# Display the updated datasets
print(put_option_data[['Close', 'SMA_10', 'SMA_50', 'EMA_10', 'EMA_50', 'RSI', 'Signal', 'Option_Strategy']].tail())
print(call_option_data[['Close', 'SMA_10', 'SMA_50', 'EMA_10', 'EMA_50', 'RSI', 'Signal', 'Option_Strategy']].tail())
```

```python
# Define parameters for risk management
capital_per_trade = 100000  # Example: 1,00,000 capital allocated per trade
max_loss_pct = 0.02  # Maximum loss per trade (2% of the capital)
stop_loss_pct = 0.10  # Stop-loss level (10% drop in option price)
target_profit_pct = 0.30  # Target profit (30% gain in option price)

# Function to calculate position size based on stop-loss and risk tolerance
def calculate_position_size(entry_price, stop_loss_price, capital_per_trade, max_loss_pct):
    loss_per_unit = entry_price - stop_loss_price  # Loss per unit option if stop-loss is hit
    position_size = (capital_per_trade * max_loss_pct) / loss_per_unit  # Number of contracts to buy
    return position_size

# Add columns to track stop-loss, target profit, and position size for both Put and Call options
put_option_data['Position_Size'] = None
put_option_data['Stop_Loss'] = None
put_option_data['Target_Profit'] = None

call_option_data['Position_Size'] = None
call_option_data['Stop_Loss'] = None
call_option_data['Target_Profit'] = None

# Iterate through the Put Option dataset and apply risk management for each trade
for i in range(len(put_option_data)):
    entry_price = put_option_data.iloc[i]['Close']
    if put_option_data.iloc[i]['Option_Strategy'] in ['Bull Put Spread', 'Bear Put Spread', 'Long Put', 'Straddle', 'Strangle']:
        # Calculate stop-loss and target profit for strategies
        if put_option_data.iloc[i]['Option_Strategy'] == 'Buy Call (Bull Call Spread)':
            stop_loss_price = entry_price * (1 - stop_loss_pct)
            target_price = entry_price * (1 + target_profit_pct)
        else:
            stop_loss_price = entry_price * (1 + stop_loss_pct)
            target_price = entry_price * (1 - target_profit_pct)

        position_size = calculate_position_size(entry_price, stop_loss_price, capital_per_trade, max_loss_pct)

        # Update columns with calculated values
        put_option_data.at[i, 'Position_Size'] = position_size
        put_option_data.at[i, 'Stop_Loss'] = stop_loss_price
        put_option_data.at[i, 'Target_Profit'] = target_price

# Iterate through the Call Option dataset and apply risk management for each trade
for i in range(len(call_option_data)):
    entry_price = call_option_data.iloc[i]['Close']
    if call_option_data.iloc[i]['Option_Strategy'] in ['Bull Call Spread', 'Bear Put Spread', 'Long Call', 'Straddle', 'Strangle']:
        # Calculate stop-loss and target profit for strategies
        if call_option_data.iloc[i]['Option_Strategy'] == 'Bull Call Spread':
            stop_loss_price = entry_price * (1 - stop_loss_pct)
            target_price = entry_price * (1 + target_profit_pct)
        else:
            stop_loss_price = entry_price * (1 + stop_loss_pct)
            target_price = entry_price * (1 - target_profit_pct)

        position_size = calculate_position_size(entry_price, stop_loss_price, capital_per_trade, max_loss_pct)

        # Update columns with calculated values
        call_option_data.at[i, 'Position_Size'] = position_size
        call_option_data.at[i, 'Stop_Loss'] = stop_loss_price
        call_option_data.at[i, 'Target_Profit'] = target_price

# Display the updated datasets with risk management info
print(put_option_data[['Close', 'SMA_10', 'SMA_50', 'EMA_10', 'EMA_50', 'RSI', 'Signal', 'Option_Strategy', 'Position_Size', 'Stop_Loss', 'Target_Profit']].tail())
print(call_option_data[['Close', 'SMA_10', 'SMA_50', 'EMA_10', 'EMA_50', 'RSI', 'Signal', 'Option_Strategy', 'Position_Size', 'Stop_Loss', 'Target_Profit']].tail())
```
